```python
import pandas as pd
```

```python
ratings = pd.read_csv('/content/drive/MyDrive/recommendation/data/ratings.csv')
ratings.sample()
```

###timestamp 컬럼 처리

```python
from datetime import datetime
datetime.fromtimestamp(845651650)
```

```python
ratings['timestamp'] = ratings['timestamp'].apply(lambda x: datetime.fromtimestamp(x))
```

```python
ratings.sample()
```

###내 데이터 붙이기

```python
myratings = pd.read_csv('/content/drive/MyDrive/recommendation/data/my-ratings_.csv')
myratings.sample()
```

```python
myratings['timestamp']=pd.to_datetime(myratings['timestamp'])
```

```python

ratings_concat = pd.concat([ratings,myratings])
ratings_concat.tail(20)
```

```python
ratings_concat.to_pickle('/content/drive/MyDrive/recommendation/data/ratings_updated.p')
```
