```python
import pandas as pd
import numpy as np
import warnings
warnings.filterwarnings('ignore')
```

###Read Data

```python
movies= pd.read_csv('/content/drive/MyDrive/recommendation/data/movies.csv')
ratings= pd.read_pickle('/content/drive/MyDrive/recommendation/data/ratings_updated.p')
genres= pd.read_pickle('/content/drive/MyDrive/recommendation/data/genres.p')
```

```python
ratings.tail()
#genres
```

###Preprocessing(전처리)

```python
ratings = ratings.merge(genres, how='inner', left_on='movieId', right_index=True)
ratings.head()
```

```python
ratings = ratings.replace(0, np.nan) #0값을 null값으로 처리
ratings.sample()
```

###Train test split

```python
from sklearn.model_selection import train_test_split
```

```python
train, test = train_test_split(ratings, random_state=42, test_size=.1)
```

```python
print(train.shape)
print(test.shape)
```

###User Profile

```python
#Train 데이터를 이용한 연산
genre_cols = genres.columns
genre_cols
```

```python
for cols in genre_cols:
   train[cols] = train[cols] * train['rating']
```

###전체 데이터로 확장(전체 유저의 평점 예측과 모델 평가)

```python
from tqdm import tqdm_notebook
```

```python
predict = []
for idx, row in tqdm_notebook(test.iterrows()):
  user = row['userId']
  #user profile * item profile
  predict.append((user_profile.loc[user] * row[genre_cols]).mean())
```

```python
test['predict'] = predict
```

```python
test.loc[test['predict'].isnull(),'predict'] = train['rating'].mean() #null 데이터들을 찾아서 predict컬럼에 train의 'rating' 평균값을 넣는다.
```

```python
test[test['predict'].isnull()]  
```

###Model validation

```python
from sklearn.metrics import mean_squared_error
```

```python
#모델에 대한 평가
#간단한 예측을 했던것(전체 평균을 넣어주는것것, 특정 유저평균으로 넣어주는것, 특정 아이템 평균으로 넣어주는 것)보다는 
#개인의 프로파일을 기반으로해서 추천을 해주는것이 조금 더 정확하다는것을 알 수 있다.
mse = mean_squared_error(test['rating'], test['predict'])
rmse = np.sqrt(mse)
rmse
```
